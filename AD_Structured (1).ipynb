```python
"""
Benchmark 3: Tabular-Only Deep Neural Network
Tests whether deep learning on clinical features alone helps

FIXES (minimal — ablation study only):
- StandardScaler fitted on train split only, applied to val
- Input sequence truncated to visits up to and including MCI (no post-MCI leakage)
- engineer_features uses causal computation (no .diff() / .rolling() look-ahead)
- adas_mmse_discordance uses per-visit values only (no global mean/std)
- concordance_index wrapped in try/except for small val sets
"""

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import numpy as np
import pickle
from sklearn.preprocessing import StandardScaler
from lifelines.utils import concordance_index
import warnings
warnings.filterwarnings('ignore')

CONFIG = {
    'tab_hidden': 256,
    'lstm_hidden': 128,
    'lstm_layers': 2,
    'dropout': 0.3,
    'lr': 5e-4,
    'weight_decay': 1e-4,
    'epochs': 100,
    'batch_size': 16,
    'max_seq_len': 10,
    'device': 'cuda' if torch.cuda.is_available() else 'cpu'
}

# ============================================================================
# FEATURE ENGINEERING
# ============================================================================

STATIC_FEATURES = [
    'age_bl', 'PTGENDER_encoded', 'PTEDUCAT', 'PTETHCAT_encoded',
    'PTRACCAT_encoded', 'PTMARRY_encoded'
]

TEMPORAL_FEATURES = [
    'time_from_baseline', 'AGE', 'age_since_bl', 'mmse_slope',
    'adas13_slope', 'dx_progression', 'cog_decline_index',
    'visit_number', 'MMSE', 'ADAS13',
    'age_mmse_interaction', 'education_cognitive_reserve', 'rapid_decline_flag',
    'mmse_severity', 'weighted_mmse_decline', 'mmse_variability', 'adas_mmse_discordance'
]

def engineer_features(df):
    """Causal feature engineering — no look-ahead."""
    df = df.copy().sort_values("Years_bl").reset_index(drop=True)

    df["time_from_baseline"] = df["Years_bl"] - df["Years_bl"].iloc[0]
    df["age_bl"]             = df["AGE"].iloc[0]
    df["age_since_bl"]       = df["AGE"] - df["age_bl"]
    df["visit_number"]       = range(len(df))

    dx_map = {"CN": 0, "MCI": 1, "AD": 2, "Dementia": 2}
    df["dx_num"] = df["DX"].map(dx_map).fillna(0)

    # Initialise causal columns
    df["mmse_slope"]            = 0.0
    df["adas13_slope"]          = 0.0
    df["dx_progression"]        = 0.0
    df["mmse_variability"]      = 0.0
    df["adas_mmse_discordance"] = 0.0

    for t in range(len(df)):
        past = df.iloc[:t + 1]
        if len(past) >= 2:
            dt = past["Years_bl"].iloc[-1] - past["Years_bl"].iloc[-2]
            if dt != 0:
                df.loc[t, "mmse_slope"]   = (past["MMSE"].iloc[-1]   - past["MMSE"].iloc[-2])   / dt
                df.loc[t, "adas13_slope"] = (past["ADAS13"].iloc[-1] - past["ADAS13"].iloc[-2]) / dt
            df.loc[t, "dx_progression"]   = past["dx_num"].iloc[-1] - past["dx_num"].iloc[-2]
            df.loc[t, "mmse_variability"]  = float(past["MMSE"].std())
        # absolute difference at current visit — no global stats
        df.loc[t, "adas_mmse_discordance"] = abs(
            float(past["ADAS13"].iloc[-1]) - float(past["MMSE"].iloc[-1])
        )

    df["cog_decline_index"]           = df["ADAS13"] - df["MMSE"]
    df["age_mmse_interaction"]        = df["AGE"] * (30 - df["MMSE"]) / 30
    df["education_cognitive_reserve"] = df["PTEDUCAT"] * df["MMSE"] / 30
    df["rapid_decline_flag"]          = (df["mmse_slope"] < -2).astype(float)
    df["weighted_mmse_decline"]       = df["mmse_slope"] * np.exp(-0.1 * df["time_from_baseline"])

    mmse_bins = [0, 20, 24, 30]
    df["mmse_severity"] = pd.cut(df["MMSE"], bins=mmse_bins, labels=[2, 1, 0]).astype(float)

    df = df.fillna(0)
    return df


# ============================================================================
# DATASET
# ============================================================================

class TabularOnlyDataset(Dataset):
    def __init__(self, manifest, valid_patients, max_seq_len=10):
        self.sequences   = []
        self.max_seq_len = max_seq_len
        self.scaler      = None  # set via apply_scaler() after split

        manifest = manifest.copy()
        manifest["path"] = manifest["path"].str.replace("\\", "/", regex=False)
        manifest["path"] = "./AD_Multimodal/TFN_AD/" + manifest["path"]

        processed         = 0
        skipped_not_valid = 0

        for ptid in manifest['PTID'].unique():
            if ptid not in valid_patients:
                skipped_not_valid += 1
                continue

            try:
                patient_rows = manifest[manifest['PTID'] == ptid]
                if len(patient_rows) == 0:
                    continue

                df = pd.read_pickle(patient_rows.iloc[0]["path"])
                df = df.sort_values("Years_bl").reset_index(drop=True)

                dx_seq = df["DX"].tolist()
                if "MCI" not in dx_seq:
                    continue

                mci_idx  = dx_seq.index("MCI")
                ad_idx   = next(
                    (i for i, x in enumerate(dx_seq[mci_idx + 1:], start=mci_idx + 1)
                     if x in ["AD", "Dementia"]),
                    -1
                )
                mci_time = float(df["Years_bl"].iloc[mci_idx])

                if ad_idx != -1:
                    time_to_event = float(df["Years_bl"].iloc[ad_idx]) - mci_time
                    event = 1
                else:
                    time_to_event = float(df["Years_bl"].iloc[-1]) - mci_time
                    event = 0

                # Only feed visits up to and including MCI — no post-MCI leakage
                df_obs = df.iloc[:mci_idx + 1].copy().reset_index(drop=True)
                df_obs = engineer_features(df_obs)

                if len(df_obs) < 1:
                    continue

                tabs, times = [], []
                valid_visits = 0

                for _, visit in df_obs.iterrows():
                    tabs.append(visit[TEMPORAL_FEATURES + STATIC_FEATURES].values.astype(np.float32))
                    times.append(float(visit["Years_bl"]))
                    valid_visits += 1
                    if valid_visits >= max_seq_len:
                        break

                pad_len = max_seq_len - valid_visits
                for _ in range(pad_len):
                    tabs.append(np.zeros_like(tabs[-1]))
                    times.append(times[-1])

                self.sequences.append({
                    'ptid':          ptid,
                    'tabs':          np.array(tabs,  dtype=np.float32),
                    'times':         np.array(times, dtype=np.float32),
                    'seq_len':       valid_visits,
                    'time_to_event': time_to_event,
                    'event':         event,
                })
                processed += 1

            except Exception as e:
                continue

        print(f"  Processed: {processed} valid patients")
        print(f"  Skipped (not in valid set): {skipped_not_valid}")

    def apply_scaler(self, scaler):
        self.scaler = scaler
        for seq in self.sequences:
            real_len = seq['seq_len']
            seq['tabs'][:real_len] = scaler.transform(seq['tabs'][:real_len])
            if real_len < len(seq['tabs']):
                seq['tabs'][real_len:] = 0.0

    def __len__(self):
        return len(self.sequences)

    def __getitem__(self, idx):
        seq = self.sequences[idx]
        return (
            seq['tabs'], seq['times'], seq['seq_len'],
            seq['time_to_event'], seq['event'], seq['ptid']
        )


# ============================================================================
# MODEL (unchanged from original)
# ============================================================================

class TabularOnlyModel(nn.Module):
    def __init__(self, tab_dim, config):
        super().__init__()

        self.tab_encoder = nn.Sequential(
            nn.Linear(tab_dim, config['tab_hidden']),
            nn.BatchNorm1d(config['tab_hidden']),
            nn.ReLU(),
            nn.Dropout(config['dropout']),
            nn.Linear(config['tab_hidden'], config['tab_hidden'] // 2),
            nn.BatchNorm1d(config['tab_hidden'] // 2),
            nn.ReLU(),
            nn.Dropout(config['dropout']),
            nn.Linear(config['tab_hidden'] // 2, 128),
            nn.ReLU()
        )

        self.lstm = nn.LSTM(
            input_size=128,
            hidden_size=config['lstm_hidden'],
            num_layers=config['lstm_layers'],
            batch_first=True,
            dropout=config['dropout'] if config['lstm_layers'] > 1 else 0,
            bidirectional=True
        )

        self.risk_head = nn.Sequential(
            nn.Linear(config['lstm_hidden'] * 2, 64),
            nn.ReLU(),
            nn.Dropout(config['dropout']),
            nn.Linear(64, 1)
        )

    def encode_tabular(self, tab):
        return self.tab_encoder(tab)

    def forward(self, tab_seq, seq_lengths):
        batch_size, seq_len, _ = tab_seq.shape
        tab_feats = torch.stack(
            [self.encode_tabular(tab_seq[:, t]) for t in range(seq_len)],
            dim=1
        )
        seq_lengths_clamped = seq_lengths.cpu().clamp(min=1)
        packed = nn.utils.rnn.pack_padded_sequence(
            tab_feats, seq_lengths_clamped, batch_first=True, enforce_sorted=False
        )
        _, (h_n, _) = self.lstm(packed)
        h_final = torch.cat([h_n[-2], h_n[-1]], dim=1)
        return self.risk_head(h_final)


# ============================================================================
# LOSS & TRAINING (unchanged from original)
# ============================================================================

def cox_loss(risk_scores, times, events):
    order = torch.argsort(times, descending=True)
    risk_scores = risk_scores[order]
    events      = events[order]
    log_risk    = risk_scores.view(-1)
    log_cumsum_hazard = torch.logcumsumexp(log_risk, dim=0)
    loss = -(log_risk - log_cumsum_hazard) * events
    return loss.sum() / (events.sum() + 1e-7)


def train_epoch(model, loader, optimizer, device):
    model.train()
    total_loss = 0
    for batch in loader:
        tabs, times, seq_lens, t_event, event, _ = batch
        tabs     = torch.FloatTensor(tabs).to(device)
        seq_lens = torch.LongTensor(seq_lens)
        t_event  = t_event.float().to(device)
        event    = event.float().to(device)

        risk_scores = model(tabs, seq_lens)
        loss = cox_loss(risk_scores.squeeze(), t_event, event)

        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        total_loss += loss.item()

    return total_loss / len(loader)


def validate(model, loader, device):
    model.eval()
    all_risks, all_times, all_events = [], [], []
    with torch.no_grad():
        for batch in loader:
            tabs, times, seq_lens, t_event, event, _ = batch
            tabs     = torch.FloatTensor(tabs).to(device)
            seq_lens = torch.LongTensor(seq_lens)
            risks    = model(tabs, seq_lens)
            all_risks.extend(risks.cpu().numpy().flatten())
            all_times.extend(t_event.numpy())
            all_events.extend(event.numpy())

    try:
        return concordance_index(
            np.array(all_times),
            -np.array(all_risks),
            np.array(all_events).astype(bool)
        )
    except ZeroDivisionError:
        return 0.5


# ============================================================================
# EXPORT (unchanged from original)
# ============================================================================

def export_features(model, loader, device, output_path):
    model.eval()
    rows = []
    with torch.no_grad():
        for batch in loader:
            tabs, times, seq_lens, t_event, event, ptids = batch
            tabs     = torch.FloatTensor(tabs).to(device)
            seq_lens = torch.LongTensor(seq_lens)
            for i in range(len(ptids)):
                tab_feat = model.encode_tabular(tabs[i:i + 1, 0])
                row = {
                    "PTID":          ptids[i],
                    "Years_bl":      float(times[i][0]),
                    "time_to_event": float(t_event[i]),
                    "event":         int(event[i]),
                }
                for k, val in enumerate(tab_feat[0].cpu().numpy()):
                    row[f"tab_feat_{k}"] = float(val)
                rows.append(row)

    df = pd.DataFrame(rows).sort_values(['PTID', 'Years_bl'])
    df.to_csv(output_path, index=False)
    print(f"\n✓ Exported to {output_path}")
    print(f"  Patients: {df['PTID'].nunique()}, Features: {len([c for c in df.columns if c.startswith('tab_feat')])}")
    return df


def export_patient_level_tabular(model, loader, device, output_path):
    """
    Export patient-level summary for TabularOnlyModel.
    Uses a forward hook on risk_head to capture z_final (LSTM output).
    Also computes z_slope from per-visit tabular encodings.
    
    Call in main() after export_features():
        export_patient_level_tabular(model, full_loader, device, "tabular_patient_level.csv")
    """
    model.eval()
    rows = []
 
    with torch.no_grad():
        for batch in loader:
            tabs, times, seq_lens, t_event, event, ptids = batch
            tabs       = torch.FloatTensor(tabs).to(device)
            times_np   = times.numpy()          # keep CPU copy for slope calc
            seq_lens_t = torch.LongTensor(seq_lens)
 
            # --- Capture z_final via hook on risk_head input ---
            captured = []
            def _hook(module, inp, out):
                captured.append(inp[0].detach().cpu())
            hook = model.risk_head[0].register_forward_hook(_hook)   # first Linear in risk_head
 
            risk_scores = model(tabs, seq_lens_t)
            hook.remove()
 
            z_final_batch = captured[0] if captured else None   # (bs, lstm_hidden*2)
 
            # --- Per-visit tabular encoding for slope calculation ---
            bs, seq_len, _ = tabs.shape
            visit_encs = []
            for t in range(seq_len):
                enc = model.encode_tabular(tabs[:, t])   # (bs, 128)
                visit_encs.append(enc.cpu().numpy())
            visit_encs = np.array(visit_encs)   # (seq_len, bs, 128)
 
            for i in range(len(ptids)):
                slen   = int(seq_lens[i])
                row = {
                    "PTID":          ptids[i],
                    "time_to_event": float(t_event[i]),
                    "event":         int(event[i]),
                    "risk_score":    float(risk_scores[i].cpu()),
                    "Years_bl":      float(times_np[i, slen - 1]),
                    "n_visits":      slen,
                }
 
                # z_final features
                if z_final_batch is not None:
                    for k, val in enumerate(z_final_batch[i].numpy()):
                        row[f"z_final_{k}"] = float(val)
                else:
                    # Fallback: use last-visit tabular encoding
                    for k, val in enumerate(visit_encs[slen - 1, i]):
                        row[f"z_final_{k}"] = float(val)
 
                # z_slope: linear slope of per-visit encoding across real visits
                real_encs  = visit_encs[:slen, i, :]   # (slen, 128)
                visit_t    = times_np[i, :slen]
                dt         = visit_t - visit_t[0]
 
                if slen >= 2 and dt[-1] > 0:
                    for k in range(real_encs.shape[1]):
                        cov_val = np.cov(dt, real_encs[:, k])[0, 1]
                        slope   = cov_val / (np.var(dt) + 1e-8)
                        row[f"z_slope_{k}"] = float(slope)
                else:
                    for k in range(real_encs.shape[1]):
                        row[f"z_slope_{k}"] = 0.0
 
                rows.append(row)
            captured.clear()
 
    df = pd.DataFrame(rows).sort_values("PTID").reset_index(drop=True)
    df.to_csv(output_path, index=False)
    n_zf = len([c for c in df.columns if c.startswith("z_final_")])
    n_zs = len([c for c in df.columns if c.startswith("z_slope_")])
    print(f"\n✓ Tabular patient-level export: {len(df)} patients")
    print(f"  z_final: {n_zf}  z_slope: {n_zs}")
    return df


# ============================================================================
# MAIN
# ============================================================================

def main():
    print("=" * 80)
    print("BENCHMARK 3: TABULAR-ONLY DEEP NETWORK")
    print("=" * 80)

    device = CONFIG['device']
    print(f"\nDevice: {device}")

    print("\nLoading valid patient list...")
    with open('VALID_PATIENTS.pkl', 'rb') as f:
        VALID_PATIENTS = pickle.load(f)
    VALID_PATIENTS = set(str(p) for p in VALID_PATIENTS)
    print(f"Valid patients: {len(VALID_PATIENTS)}")

    manifest = pd.read_csv("./AD_Multimodal/TFN_AD/AD_Patient_Manifest.csv")
    manifest["PTID"] = manifest["PTID"].astype(str)

    # Patient-level split BEFORE dataset creation
    all_ptids   = sorted(list(set(manifest["PTID"]) & VALID_PATIENTS))
    np.random.seed(42)
    train_ptids = set(np.random.choice(all_ptids, size=int(0.8 * len(all_ptids)), replace=False))
    val_ptids   = set(all_ptids) - train_ptids

    print(f"\nTrain patients: {len(train_ptids)}")
    print(f"Val patients  : {len(val_ptids)}")

    print("\nBuilding TRAIN dataset...")
    train_dataset = TabularOnlyDataset(manifest, train_ptids, max_seq_len=CONFIG['max_seq_len'])
    print("\nBuilding VAL dataset...")
    val_dataset   = TabularOnlyDataset(manifest, val_ptids,   max_seq_len=CONFIG['max_seq_len'])

    # Fit scaler on train only
    train_tabs = np.vstack([s['tabs'][:s['seq_len']] for s in train_dataset.sequences])
    scaler = StandardScaler()
    scaler.fit(train_tabs)
    train_dataset.apply_scaler(scaler)
    val_dataset.apply_scaler(scaler)
    print("\nScaler fitted on train only — no leakage")

    train_loader = DataLoader(train_dataset, batch_size=CONFIG['batch_size'], shuffle=True,  num_workers=0)
    val_loader   = DataLoader(val_dataset,   batch_size=CONFIG['batch_size'], shuffle=False, num_workers=0)

    tab_dim = train_dataset.sequences[0]['tabs'].shape[1]
    print(f"Tabular features: {tab_dim}")

    model     = TabularOnlyModel(tab_dim, CONFIG).to(device)
    optimizer = torch.optim.AdamW(model.parameters(), lr=CONFIG['lr'], weight_decay=CONFIG['weight_decay'])
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=5)

    print("\nTraining...")
    best_c_index     = 0
    patience_counter = 0

    for epoch in range(CONFIG['epochs']):
        train_loss  = train_epoch(model, train_loader, optimizer, device)
        val_c_index = validate(model, val_loader, device)
        scheduler.step(val_c_index)

        print(f"Epoch {epoch+1}/{CONFIG['epochs']} - Loss: {train_loss:.4f}, C-index: {val_c_index:.4f}")

        if val_c_index > best_c_index:
            best_c_index = val_c_index
            torch.save(model.state_dict(), 'tabular_only_model.pth')
            patience_counter = 0
        else:
            patience_counter += 1

        if patience_counter >= 15:
            print("Early stopping")
            break

    model.load_state_dict(torch.load('tabular_only_model.pth', weights_only=False))

    print("\nRebuilding full dataset for export...")
    full_dataset = TabularOnlyDataset(manifest, VALID_PATIENTS, max_seq_len=CONFIG['max_seq_len'])
    full_dataset.apply_scaler(scaler)
    full_loader  = DataLoader(full_dataset, batch_size=CONFIG['batch_size'], shuffle=False, num_workers=0)

    export_features(model, full_loader, device, "tabular_only_features.csv")
    export_patient_level_tabular(model, full_loader, device, "tabular_patient_level.csv")

    print("\n" + "=" * 80)
    print(f"✓ BEST C-INDEX: {best_c_index:.4f}")
    print("=" * 80)


if __name__ == "__main__":
    main()
```
